```python
window_size = len(OSAdata[1])
width = 2
height = -40
distance = 10
lambdas, intensities, delta_lambdas = osam.find_peak_spacing(OSAdata[0], OSAdata[1], window_size, width=width, height=height, distance=distance, filename=filename, save=False, target_dir=target_dir)
```

```python
osax = OSAdata[0]
osay = OSAdata[1]
peak_locations = lambdas

def fit_lasing_mode(osax, osay, peak_locations, save = False):
    indlocs = []
    for i,j in enumerate(osax):
        if j in peak_locations:
            indlocs.append(i)
    print(indlocs)

    # y_crop = min(osay) + np.abs(max(osay) - min(osay))
    # osay = [y for y in osay if y >= y_crop]
    # osay = list(np.array(osay) - min(osay))

    # plt.plot(osax, osay)
    # plt.show()
    
    local = fit.get_local_indices(indlocs, len(osay))
    print(local)

    for i, loc in enumerate(indlocs):
        print(min(local[i]), max(local[i]))
        print(i, loc)
        plt.plot(osax[int(min(local[i])):int(max(local[i]))], osay[int(min(local[i])):int(max(local[i]))])
        plt.show()
        lorentz_coeff = osam.odr_auto_lorentz(osax[int(min(local[i])):int(max(local[i]))], osay[int(min(local[i])):int(max(local[i]))], int(loc-min(local[i])))
        # lorentz_coeff = odr_auto_lorentz(osax, osay, int(loc))
        lorentz_fit = fit.odr_lorentzian_lineshape(lorentz_coeff, osax)

        print('Lorentz fit coefficients (A, mu, gamma, offset = ', lorentz_coeff)
        linewidth = tools.get_linewidth(lorentz_coeff[1], lorentz_coeff[2])
        plt.plot(osax, lorentz_fit, label=r'$\delta\omega_{}=$'.format(i+1)+str(round(linewidth*10**-9,2))+' GHz', linewidth=0.8, zorder=3, color='tab:blue')

    plt.plot(osax, osay, label = 'OSA signal', color = 'lightgrey', linewidth = 0.5)

    plt.ylim(min(osay), max(osay))
    plt.grid(axis = 'x', which = 'both')
    plt.legend()
    plt.xlabel('Wavelength (nm)')
    plt.ylabel('Intensity (a.u.)')
    title = 'OSA Spectrum Lasing Modes Lorentz Fit'
    plt.title(title, fontsize = 8)
    if save is True:
        plt.savefig(target_dir+str(title)+"_plot.png", bbox_inches='tight',pad_inches=0.0, format = 'png', dpi=1200)
    plt.show()

fit_lasing_mode(osax, osay, peak_locations, save = False)
```
